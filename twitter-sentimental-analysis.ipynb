```python
import re
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
import nltk
nltk.download('stopwords')
```

```python
print(stopwords.words('english'))
```

```python
column_names = ['target', 'id', 'date', 'flag', 'user', 'text']
twitter_data = pd.read_csv('/kaggle/input/sentiment140/training.1600000.processed.noemoticon.csv',
                          names=column_names ,encoding = 'ISO-8859-1')
```

```python
twitter_data.replace({'target': {4:1}}, inplace=True)
```

```python
twitter_data['target'].value_counts()
```

```python
port_stem = PorterStemmer()
```

```python
def stemming(content):

    stemmed_content = re.sub('[^a-zA-Z]',' ', content)
    stemmed_content = stemmed_content.lower()
    stemmed_content = stemmed_content.split()
    stemmed_content = [port_stem.stem(word) for word in stemmed_content if not word in 
                       stopwords.words('english')]
    stemmed_content = ' '.join(stemmed_content)

    return stemmed_content
```

```python
twitter_data['stemmed_content'] = twitter_data['text'].apply(stemming)
```

```python
twitter_data.head()
```

```python
print(twitter_data['stemmed_content'])
```

```python
print(twitter_data['target'])
```

```python
X = twitter_data['stemmed_content'].values
Y = twitter_data['target'].values
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
print(X_train)
```

```python
print(X_test)
```

```python
vectorizer = TfidfVectorizer()

X_train = vectorizer.fit_transform(X_train)
X_test = vectorizer.transform(X_test)
```

```python
print(X_train)
```

```python
print(X_test)
```

```python
model = LogisticRegression(max_iter=1000)
model.fit(X_train, Y_train)
```

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
```

```python
print("Accuracy score on training data :", training_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)
```

```python
print("Accuracy score on test data :", test_data_accuracy)
```
